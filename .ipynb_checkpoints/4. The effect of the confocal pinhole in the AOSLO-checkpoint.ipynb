## 4. The effect of the confocal pinhole in the AOSLO

In the previous notebook, we convolved a psf measured from a real eye with a simulated retinal image.  However, in the AOSLO this is not so simple, as:

1. The system is double-pass - so we illuminate the retina and then collect light returning from the retina
2. The system scans a point of light across the retina 
3. The collected light passes through a confocal pinhole to photodetector (APD or PMT)


In this notebook, we will go over all the steps required to generate an image at the detector.

There are three computations that must be performed:
1. The input PSF must be multiplied by the synthetic retinal mosaic, giving the reflectance profile of the illuminated patch of retina
2. The illuminated patch should be convolved with the output PSF to replicate reimaging it on to the pinhole
3. The image formed at the pinhole should be multiplied by the transmission function defining the pinhole.


```python
import numpy
import matplotlib.pylab as plt
from tqdm import tqdm, trange, tqdm_notebook
from astropy.io import fits
%matplotlib inline
import aotools
import correlation_convolution
from scipy.ndimage.interpolation import shift
from IPython.display import HTML
import PIL.Image as Image

import scipy.signal
j = 0+1j
```

```python
fitsdata = fits.open('tabai5mm.fits')
data=fitsdata[0].data

wavelength=833e-3 #(this is the wavelength data was taken at in microns)
wavelength_m=833e-9 #(wavelength in metres - we'll use this later)
diameter=5e-3 #(pupil diameter in m)

data[:,:,0:2]=0 #remove piston, tip and tilt again

array_size=128
oversampling = 2
```

## Step 1. Illuminate a patch on the retina from a point source of light directed through the eye's optics

Lets import our retinal mosaic from ERICA again

```python
image = Image.open('foveatedretina.bmp')
image = numpy.array(image) 
plt.figure(figsize=(8,8))
plt.imshow(image,cmap='gray')
```

Generate a PSF again for the first eye at time 0

```python
firsteye_t0 = (data[0,0,:])
z=aotools.zernikeArray(65,array_size)
zflat=numpy.reshape(z, (z.shape[0],z.shape[1]*z.shape[2]))
wavefront = numpy.dot(firsteye_t0,zflat)
wavefront = numpy.reshape(wavefront, (z.shape[1],z.shape[2]))
wavefront *= aotools.circle(array_size/2, array_size)
wavefront_radians = wavefront*((2*numpy.pi)/wavelength)

padded_pupil=numpy.zeros((int(oversampling*array_size),int(oversampling*array_size)),dtype=numpy.complex64)
padded_pupil[:array_size,:array_size]=numpy.exp(j*wavefront_radians)

psf = numpy.abs(aotools.ft2(padded_pupil, delta=diameter/array_size))**2

plt.imshow(psf)
```

We need to determine how much of the retinal image to include in the convolution with the PSF, based on how broad the PSF is

```python
psf_norm = (psf/psf.max())
y,x = numpy.indices((psf_norm.shape))
r = numpy.sqrt((x/2)**2+(y/2)**2) #radius of psf
r = r.astype(int)
tbin = numpy.bincount(r.ravel(),psf_norm.ravel())
nr=numpy.bincount(r.ravel())
radialprofile = tbin/nr
plt.plot(radialprofile)
#bincount counts number of occurrences of each value in an array of integers.  
threshold = 1.0/array_size
plt.plot(numpy.arange(0,radialprofile.shape[0]),numpy.ones(radialprofile.shape[0])*threshold)
box_size = numpy.where(radialprofile>threshold)[0].max()*2
print('Crop image to %i by %i pixels'%(box_size,box_size))
```

Crop the centre of the psf

```python
psf_box = shift(psf_norm,(-psf_norm.shape[0]/2.+box_size*0.5,-psf_norm.shape[1]/2.+box_size*0.5))[:int(round(box_size)),:int(round(box_size))]
plt.imshow(psf_box)
```

Crop the image and multiply by the psf

```python
Image_crop = image[:box_size,:box_size]
plt.imshow(Image_crop)
```

```python
illuminated_patch = psf_box*Image_crop
plt.imshow(illuminated_patch)
```

Lets zoom in on all these steps

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

ax1.imshow(Image_crop[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax2.imshow(psf_box[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax3.imshow(illuminated_patch[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])

ax1.set_title('retinal mosaic')
ax2.set_title('psf')
ax3.set_title('illuminated patch')
ax1.set_axis_off()
ax2.set_axis_off()
ax3.set_axis_off()
```

So this is an example of what the illuminated patch of retina looks like at one point in the raster scan (with no AO correction)

## 2. Convolve the patch with the output PSF to replicate reimaging it on to the pinhole

We've already done this in notebook 3! - let's do a convolution

We'll assume that the output PSF is the same as the input PSF.

```python
image_at_pinhole = scipy.signal.convolve2d(illuminated_patch,psf_box,mode='same')
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

ax1.imshow(illuminated_patch[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax2.imshow(psf_box[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax3.imshow(image_at_pinhole[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])

ax1.set_title('illuminated patch')
ax2.set_title('psf')
ax3.set_title('image at pinhole')
ax1.set_axis_off()
ax2.set_axis_off()
ax3.set_axis_off()
```

## Step 3. Multiply the image formed at the pinhole by the transmission function defining the pinhole


We need to define the pinhole first.  Let's say our pinhole is 2 Airy disc diameters (2ADD) - this is typically used in the AOSLO.  


```python
ADD = (2.44 * wavelength_m*180*60/(diameter*numpy.pi))
print ('ADD = %.3f arcmin'%ADD)
```

```python
psf_pixel_scale = (1/(oversampling*diameter))*(wavelength_m*180*60/numpy.pi)
print ('PSF pixel scale = %.3f arcmin'%psf_pixel_scale)
```

```python
PH_size_pixels=(2*ADD)/psf_pixel_scale
print(PH_size_pixels)
```

```python
pinhole_add = 2
ph_scale_up = 100.
pinhole_radius_pixels = pinhole_add * oversampling * 1.22

cenx_box = (psf_box.shape[1])/2.
ceny_box = (psf_box.shape[0])/2.
y, x = numpy.ogrid[0:psf0_box.shape[0]-1:psf0_box.shape[0]*ph_scale_up*1j, 0:psf0_box.shape[1]-1:psf0_box.shape[1]*ph_scale_up*1j]
r = numpy.sqrt((x-cenx_box)**2 + (y-ceny_box)**2)
pinhole_scale_up = numpy.where(r<=pinhole_radius_pixels, 1., 0.)
pinhole0 = numpy.asarray(Image.fromarray(pinhole_scale_up).resize(psf0_box.shape, Image.ANTIALIAS))
pinhole_box = numpy.clip(pinhole0,0.,1.0)
```

```python
plt.imshow(pinhole_box[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
```

This is our pinhole.  Multiply this by the image at the pinhole to get the image obtained at the detector

```python
image_at_detector = image_at_pinhole*pinhole_box
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

ax1.imshow(image_at_pinhole[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax2.imshow(pinhole_box[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])
ax3.imshow(image_at_detector[int(box_size/3):int(2*box_size/3),int(box_size/3):int(2*box_size/3)])

ax1.set_title('illuminated patch')
ax2.set_title('psf')
ax3.set_title('image at pinhole')
ax1.set_axis_off()
ax2.set_axis_off()
ax3.set_axis_off()
```
