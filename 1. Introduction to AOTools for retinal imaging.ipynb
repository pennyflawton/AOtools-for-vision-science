## Imaging Through the Optics of the Eye

This is a workshop designed to show how aotools can be used not only to study aberrations in astronomy, but also in ophthalmology and vision science.

(Note that in this workshop we only study monochromatic aberrations in the eye)

After completing this workshop, you should be able to:

- Create an aberrated PSF using real data from a dataset of human eyes  𝑎
 
- Apply this to an extended image of retinal cells (simulated using ERICA)  𝑏
